# Vamos ver mais exemplos com IF?

```python
x = 5
if x > 10:
    print("Maior que 10")
else:
    print("Menor que 10")
```

```python
nome = "Maria"
if nome == "Allan":
    print("Seja bem-vindo, Allan")
else:
    print("Usuário desconhecido")
```

```python
dia = 3
if dia == 1:
    print("Segunda")
elif dia == 2:
    print("Terça")
elif dia == 3:
    print("Quarta")
else:
    print("Outro dia")
```

```python
cor = "verde"
if cor == "vermelho":
    print("Pare")
elif cor == "amarelo":
    print("Atenção")
elif cor == "verde":
    print("Siga")
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(2, 10, 2):
    print(i)
```

```python
for palavra in ["sol"]:
    print(palavra.upper())
```

```python
for letra in ["a", "e", "i", "o", "u"]:
    print(f"Vogal: {letra}")
```

```python
x = 0
while x < 5:
    print(x)
    x += 1
```

```python
n = 5
while n >= 0:
    print(n)
    n -= 1
```

```python
x = 1
while x <= 3:
    print("Loop externo:", x)
    y = 1
    while y <= 2:
        print("Loop interno:", y)
        y += 1
    x += 1
```

```python
palavra = "Python"
i = 0
while i < len(palavra):
    print(palavra[i])
    i += 1
```

```python
seq = "AUGCUA"
if "U" in seq:
    print("Sequência de RNA")
else:
    print("Sequência de DNA")
```

```python
seq = "ATGCGTACGT"
for i in range(0, len(seq), 3):
    print("Códon:", seq[i:i+3])
```

```python
seq = "AUGCCCUAAUGA"
i = 0
while i < len(seq):
    codon = seq[i:i+3]
    if codon in ["UAA", "UAG", "UGA"]:
        print("Stop códon encontrado:", codon)
        break
    i += 3
```
